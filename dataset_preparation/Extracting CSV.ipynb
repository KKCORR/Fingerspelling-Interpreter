```python
# function for get file name from directory
def getFileList(directory):
  file_list = []
  # *.* use for get every file type in directory
  for file in glob.glob(dataset_url + directory + "/*.*"):
    # print(file)
    file_list.append(file)
  return file_list

# function call mediapipe for static image
# input is directory from dataset folder onward, eg. /22_03_2021/1
# output store in output_dataset with the same folder
def extractCSV(directory):
  file_list = getFileList(directory)
  mp_drawing = mp.solutions.drawing_utils
  mp_hands = mp.solutions.hands
  row_list = [['number']]
  for i in range(21):
    row_list[0].append(str(i) + 'x')
    row_list[0].append(str(i) + 'y')
    row_list[0].append(str(i) + 'z')
  with mp_hands.Hands(
      static_image_mode=True,
      max_num_hands=2,
      min_detection_confidence=0.5) as hands:
    for idx, file in enumerate(file_list):
      # Read an image, flip it around y-axis for correct handedness output (see
      # above).
      image = cv2.flip(cv2.imread(file), 1)
      # Convert the BGR image to RGB before processing.
      results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

      # Print handedness and draw hand landmarks on the image.
      # print('Handedness:', results.multi_handedness)
      if not results.multi_hand_landmarks:
        continue
      image_height, image_width, _ = image.shape
      annotated_image = image.copy()
      for hand_landmark in results.multi_hand_landmarks:
        row = [idx]
        for i in range(21):
          row.append(hand_landmark.landmark[i].x)
          row.append(hand_landmark.landmark[i].y)
          row.append(hand_landmark.landmark[i].z)
      maxx = 0
      minx = 1000
      maxy = 0
      miny = 1000
      maxz = 0
      minz = 1000
      for i in range(1,len(row),3):
        if row[i] > maxx:
          maxx = row[i]
        if row[i] < minx:
          minx = row[i]
      for i in range(1,len(row),3):
        row[i] = (row[i]-minx)/(maxx-minx)
      for i in range(2,len(row),3):
        if row[i] > maxy:
          maxy = row[i]
        if row[i] < miny:
          miny = row[i]
      for i in range(2,len(row),3):
        row[i] = (row[i]-miny)/(maxy-miny)
      for i in range(3,len(row),3):
        if row[i] > maxz:
          maxz = row[i]
        if row[i] < minz:
          minz = row[i]
      for i in range(3,len(row),3):
        row[i] = (row[i]-minz)/(maxz-minz)
      row_list.append(row)
    print(row_list)
        
        # print(
        #     f'Index finger tip coordinates: (',
        #     f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * image_width}, '
        #     f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * image_height})'
        # )
    save_dir = drive_url + "/CSV" + "/" + directory[-1] + '.csv'
    if directory[-2] != "/":
      print('yes')
      save_dir = drive_url + "/CSV" + "/" + directory[-2::] + '.csv'
    print(save_dir)
      # cv2.imwrite(save_dir, cv2.flip(annotated_image, 1))
    with open(save_dir, 'w', newline='') as file:
      writer = csv.writer(file)
      writer.writerows(row_list)
```
